# Web Crawler

```python
from bs4 import BeautifulSoup
import requests
comp = []
def scraper(name):
    web = 'https://www.chinabiz.org.tw/Company/SearchList?keys=' + name + '&btnSubmit='
    response = requests.get(web)
    soup = BeautifulSoup(response.content.decode('utf8', 'ignore'), "html.parser")
    cag = soup.find_all('ul',{'class':'text_list'}, limit=1)
    if len(cag):
        comp.append(True)
    else:
        comp.append(False)
```

# Choose data from company which has subsidiary in china

```python
company_in_china = company[company['In_China']]
company_in_china
```

```python
# 贊助者為公司的總數量
total = company.Candidate.value_counts()
total
```

```python
# 贊助者其中公司有在中國設廠的數量
china_company_value = company_in_china.Candidate.value_counts()
china_company_value
```

```python
plt.figure(figsize=(10,5))
myfont = FontProperties(fname=r'msj.ttf',size=14)
sns.set(font = myfont.get_family(), font_scale = 2)
sns.set_style("whitegrid",{"font.sans-serif":['Microsoft JhengHei']})
ax = sns.barplot(china_company_value.values, china_company_value.keys())
ax.set(xlabel = "子公司在中國數量", ylabel = "候選人")
for i, v in enumerate(china_company_value):
    ax.text(v, i+ .08, str(v), color='black', fontweight='bold')
plt.show()
```
